#Day 3( Intro to Machine Learning)

```python
import pandas as pd
melb_file_path = '/content/melb_data.csv'

#read the data and store data in DataFrame title melb_data
melb_data = pd.read_csv(melb_file_path)

#print a sumaary of the data in melb_data.csv
melb_data.describe()
```

```python
#show columns of the data
melb_data.columns
```

```python
#dropna drops missing values (drops things are not 'avaliable')
melb_data = melb_data.dropna(axis=0)
```

There are different ways of selecting subsets of your data. We will focus on two.



1.   Dot anotation, which we use to select the 'prediction target'
2.   Selecting with a column list, which we use to select 'features'



#Selecting the Target Prediction

```python
y = melb_data.Price
```

#Choosing a "Features"

```python
melb_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
X = melb_data[melb_features]
```

```python
X.describe()
```

```python
X.head()
```

#Building Our Model
* Scikit-learn library (sklearn)

The steps
1.   Define: What type of model will it be? A decision tree? Or some type of model?

2.   Fit: Capture patterns from the data provided.

3.   Predict

4.   Evaluate





```python
from sklearn.tree import DecisionTreeRegressor

#Define model. Specify a number for random_state to ensure same results each run
melb_model = DecisionTreeRegressor(random_state=1)

#Fit model
melb_model.fit(X, y)
```

##comment

```python
print("We will make a prediction for the following 5 houses:")
print(X.head())
print("The predictions are ")
print(melb_model.predict(X.head()))
```

```python
melb_data.head()
```

##Model Validation

*MAE : Mean Absolute Error

Error = (actual value) - (predicetd value)

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melb_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

What we have donhe here is 'In Sample'

##Split

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

melb_model2 = DecisionTreeRegressor()
melb_model2.fit(train_X, train_y)

val_predt = melb_model2.predict(val_X)
print(mean_absolute_error(val_y, val_predt))
```
